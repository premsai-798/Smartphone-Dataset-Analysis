```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
smartphones = pd.read_csv('Downloads/smartphones.csv')
smartphones
```

```python
smartphones.info()
```

```python
smartphones['model'].unique()
```

```python
# Descriptive statistics for numerical columns
print("Descriptive Statistics:")
print(smartphones.describe())
```

```python
# Correlation matrix
print("\nCorrelation Matrix:")
print(smartphones.corr())
```

```python
# Plot 1: Bar Chart - Distribution of Smartphone Brands
def plot_brand_distribution(data):
    brand_counts = data['brand_name'].value_counts()
    plt.figure(figsize=(10, 6))
    brand_counts.plot(kind='bar', color='skyblue')
    plt.title('Distribution of Smartphone Brands')
    plt.xlabel('Brand')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
plot_brand_distribution(smartphones)
```

```python
# Plot 2: Scatter Plot - Price vs Battery Capacity
def plot_price_vs_battery(data):
    plt.figure(figsize=(10, 5))
    plt.scatter(data['battery_capacity'], data['price'], alpha=0.5, color='purple')
    plt.title('Price vs Battery Capacity')
    plt.xlabel('Battery Capacity (mAh)')
    plt.ylabel('Price (INR)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
plot_price_vs_battery(smartphones)
```

```python
# Plot 3: Heatmap - Correlation of Key Features
def plot_correlation_heatmap(data):
    # Selecting a subset of relevant features for the heatmap
    relevant_features = ['price', 'battery_capacity', 'screen_size', 'avg_rating', 'internal_memory', 'refresh_rate']
    corr_matrix = data[relevant_features].corr()
    plt.figure(figsize=(8,5))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap of Key Features')
    plt.tight_layout()
    plt.show()
plot_correlation_heatmap(smartphones)
```
